```python
from selenium import webdriver

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver', chrome_options=chrome_options)
driver =webdriver.Chrome('chromedriver', chrome_options=chrome_options)
```

```python
wd.get('https://www.amazon.in/')
searchBox=wd.find_element_by_id('twotabsearchtextbox')
searchBox.send_keys('kindle')
submitBtn=wd.find_element_by_css_selector('.nav-search-submit')
submitBtn.click()
product_list=[]
for i in range(1,3):  # 1st two page
  wd.implicitly_wait(10)
  temp_product_list=wd.find_elements_by_xpath('//span[@class="a-size-medium a-color-base a-text-normal"]')
  for product in temp_product_list:
    product_list.append(product.text)
 
  next_link=wd.find_element_by_xpath('//li[@class="a-last"]/a')
  next_link.click()

for prod in product_list:
  print(prod)
  
print(len(product_list))
```
